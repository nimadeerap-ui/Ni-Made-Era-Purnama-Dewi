Membuat array Data Nilai

```python
import numpy as np

arr = np.array( [70, 65, 85, 90, 66] )
```

Tampilkan data array

```python
print("data array: ")
print(arr)
```

Fungsi statistik pada numpy

```python
nRendah = np.min(arr)
print("Nilai Terendah: ")
print(nRendah)

nTinggi = np.max(arr)
print("Nilai Tertingi: ")
print(nTinggi)

nRata = np.mean(arr)
print("Nilai rata rata: ")
print(nRata)
```
